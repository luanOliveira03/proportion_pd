```python
import pandas as pd
```

```python
sales = pd.read_csv('vgsales.csv')

sales.head()
```

```python
sales['NA_Proportion'] = (sales['NA_Sales'] / sales['Global_Sales']) * 100
sales['EU_Proportion'] = (sales['EU_Sales'] / sales['Global_Sales']) * 100
sales['JP_Proportion'] = (sales['JP_Sales'] / sales['Global_Sales']) * 100
sales['OTHER_Proportion'] = (sales['Other_Sales'] / sales['Global_Sales']) * 100
```

```python
sales.head()
```

```python
jp_sales = sales.nlargest(10, 'JP_Proportion')

print(f'As 10 maiores vendas proporcionais do Japão: {jp_sales}' )
```

```python
import matplotlib.pyplot as plt
```

```python
sales[['EU_Sales', 'Global_Sales']].plot(kind='line', marker='o')
plt.xlabel('Quantidade')
plt.ylabel('Vendas')
plt.title('Europa x Globais')
plt.legend(['Europa', 'Global'])
plt.show()
```
